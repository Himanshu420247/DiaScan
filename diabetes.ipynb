```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/diabetes.csv')  # Replace 'your_file.csv' with the path to your CSV file

# Check the first few rows to understand the structure of the dataset
print(df.head())

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Histograms for each feature
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
for i, feature in enumerate(features, 1):
    plt.subplot(3, 3, i)
    df[feature].hist(bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()

# Scatter plot to show relationship between Glucose and BMI
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Glucose', y='BMI', hue='Outcome', palette='coolwarm')
plt.title('Scatter Plot of Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.legend(title='Outcome')
plt.show()

# Pair plot to show pairwise relationships
sns.pairplot(df, hue='Outcome', diag_kind='kde', palette='coolwarm')
plt.suptitle('Pair Plot of Features by Outcome', y=1.02)
plt.show()



# plt.tight_layout()
# plt.show()
```
